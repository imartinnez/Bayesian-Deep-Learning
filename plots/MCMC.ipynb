```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib as mpl
import numpy as np

# ── Paleta CUNEF ──────────────────────────────────────────
OR  = "#E8500A"
AM  = "#F4A261"
DK  = "#0D0D0D"
GR  = "#6B6B6B"
LG  = "#3A3A3A"
BL  = "#5BC8F5"
WH  = "#E8E8E8"

mpl.rcParams.update({
    "font.family":      "serif",
    "font.serif":       ["Computer Modern Roman", "Times New Roman", "DejaVu Serif"],
    "figure.facecolor": DK,
    "text.color":       WH,
    "axes.facecolor":   DK,
})
```

```python
# ── Datos fijos ───────────────────────────────────────────
accepted = [
    (340,300),(310,250),(290,210),(295,195),(310,190),
    (325,185),(335,182),(330,188),(320,192),(315,198),
    (320,202),(328,200),(332,196),(329,193),(325,196),
    (322,200),(325,205),(330,202),(335,198),(332,194),
]
rejected_pt = (260, 330)   # propuesta rechazada paso 4

# Nube final de convergencia
cloud = [
    (335,182),(320,188),(310,192),(318,198),(328,200),
    (332,196),(325,193),(322,200),(328,205),(333,202),
    (329,195),(325,199),(320,203),(315,197),(322,193),
    (330,190),(338,186),(342,192),(336,198),(328,194),
    (318,202),(312,206),(320,210),(330,207),(338,202),
    (343,196),(340,190),
]

def make_fig():
    fig, ax = plt.subplots(figsize=(3.8, 3.2))
    ax.set_facecolor(DK)
    ax.set_xlim(80, 620)
    ax.set_ylim(40, 400)
    ax.axis("off")
    return fig, ax

def posterior_rings(ax):
    cx, cy = 340, 210
    for r, op in [(18,.40),(32,.30),(50,.22),(70,.14),(95,.08),(125,.04)]:
        e = mpatches.Ellipse((cx, cy+10), r*3.2, r*2,
                             fc=OR, ec="none", alpha=op, zorder=1)
        ax.add_patch(e)

def axes_labels(ax):
    ax.annotate("", xy=(600,385), xytext=(85,385),
                arrowprops=dict(arrowstyle="-|>", color=LG, lw=0.8))
    ax.annotate("", xy=(85,45), xytext=(85,385),
                arrowprops=dict(arrowstyle="-|>", color=LG, lw=0.8))
    ax.text(340, 410, r"$\theta_1$", ha="center", va="top",
            fontsize=9, color=GR, style="italic")
    ax.text(72, 210, r"$\theta_2$", ha="center", va="center",
            fontsize=9, color=GR, style="italic", rotation=90)

def dot(ax, x, y, r, fc, ec, lw=1.5, zorder=5):
    c = plt.Circle((x, y), r, fc=fc, ec=ec, linewidth=lw, zorder=zorder)
    ax.add_patch(c)

def label(ax, x, y, txt, color=WH, size=8, dx=8, dy=0):
    ax.text(x+dx, y+dy, txt, fontsize=size, color=color,
            style="italic", va="center", zorder=7)

def step_title(ax, txt):
    ax.text(340, 52, txt, ha="center", va="top",
            fontsize=8.5, color=WH, fontweight="bold", zorder=8)

# ══════════════════════════════════════════════════════════
# PASO 1 — punto inicial
# ══════════════════════════════════════════════════════════
fig, ax = make_fig()
posterior_rings(ax)
axes_labels(ax)
dot(ax, 180, 340, 7, OR, WH, 1.5)
label(ax, 190, 340, r"$\theta^{(0)}$", WH)
step_title(ax, "1. Punto de partida")
ax.text(340, 360, "Prior fijado, inicio aleatorio en el espacio",
        ha="center", fontsize=7, color=GR)
plt.tight_layout(pad=0.3)
plt.savefig("images/mcmc_paso1.png", dpi=220, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()

# ══════════════════════════════════════════════════════════
# PASO 2 — propuesta θ*
# ══════════════════════════════════════════════════════════
fig, ax = make_fig()
posterior_rings(ax)
axes_labels(ax)
dot(ax, 180, 340, 7, OR, AM, 1.5)
label(ax, 180, 340, r"$\theta^{(0)}$")
ax.annotate("", xy=(310, 250), xytext=(187, 333),
            arrowprops=dict(arrowstyle="-|>", color=BL,
                            lw=1.0, linestyle="dashed"))
dot(ax, 310, 250, 7, "#0A1A2A", BL, 1.8)
label(ax, 320, 250, r"$\theta^{*}$", BL)
ax.text(280, 285, "propuesta", fontsize=7, color=BL, style="italic")
step_title(ax, "2. Proponer θ*")
ax.text(340, 360, r"Se samplea $\theta^* \sim q(\theta^*|\theta^{(t)})$",
        ha="center", fontsize=7, color=GR)
plt.tight_layout(pad=0.3)
plt.savefig("images/mcmc_paso2.png", dpi=220, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()

# ══════════════════════════════════════════════════════════
# PASO 3 — propuesta aceptada
# ══════════════════════════════════════════════════════════
fig, ax = make_fig()
posterior_rings(ax)
axes_labels(ax)
dot(ax, 180, 340, 6, DK, GR, 1.0)
ax.plot([180, accepted[0][0]], [340, accepted[0][1]],
        color=OR, lw=1.2, ls="--", dashes=(3,2), zorder=3)
dot(ax, accepted[0][0], accepted[0][1], 7, OR, AM, 1.5)
label(ax, accepted[0][0], accepted[0][1], r"$\theta^{(1)}$")
ax.text(230, 260, r"$\alpha \geq 1$ → ✓ aceptada",
        fontsize=7.5, color=WH)
step_title(ax, "3. Aceptar (mayor densidad)")
ax.text(340, 360, r"$p(\theta^*|\mathcal{D}) > p(\theta^{(t)}|\mathcal{D})$",
        ha="center", fontsize=7, color=GR)
plt.tight_layout(pad=0.3)
plt.savefig("images/mcmc_paso3.png", dpi=220, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()

# ══════════════════════════════════════════════════════════
# PASO 4 — propuesta rechazada
# ══════════════════════════════════════════════════════════
fig, ax = make_fig()
posterior_rings(ax)
axes_labels(ax)
dot(ax, 180, 340, 6, DK, GR, 1.0)
ax.plot([180, accepted[0][0]], [340, accepted[0][1]],
        color=OR, lw=1.0, ls="--", dashes=(3,2), zorder=3, alpha=0.5)
dot(ax, accepted[0][0], accepted[0][1], 7, OR, AM, 1.5)
ax.plot([accepted[0][0], rejected_pt[0]],
        [accepted[0][1], rejected_pt[1]],
        color=GR, lw=0.9, ls="--", dashes=(3,3), zorder=3)
dot(ax, rejected_pt[0], rejected_pt[1], 6, "#1A1A1A", "#888", 1.2)
label(ax, rejected_pt[0], rejected_pt[1], r"$\theta^{*}$ rechazada", "#888", dx=9)
ax.text(95, 310, r"$\alpha < 1$ → ✗ nos quedamos",
        fontsize=7.5, color="#888")
step_title(ax, "4. Rechazar (menor densidad)")
ax.text(340, 360, r"La cadena permanece en $\theta^{(t)}$",
        ha="center", fontsize=7, color=GR)
plt.tight_layout(pad=0.3)
plt.savefig("images/mcmc_paso4.png", dpi=220, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()

# ══════════════════════════════════════════════════════════
# PASO 5 — iteraciones acumuladas
# ══════════════════════════════════════════════════════════
fig, ax = make_fig()
posterior_rings(ax)
axes_labels(ax)
pts = accepted[:10]
for i in range(1, len(pts)):
    ax.plot([pts[i-1][0], pts[i][0]],
            [pts[i-1][1], pts[i][1]],
            color=OR, lw=0.9, ls="--", dashes=(3,2),
            alpha=0.5, zorder=3)
for i, (x,y) in enumerate(pts):
    dot(ax, x, y, 4 if i>0 else 6, OR, AM, 1.2 if i>0 else 1.5)
step_title(ax, "5. Iteraciones acumuladas")
ax.text(340, 360, "La cadena comienza a concentrarse acumulando\nmuestras...",
        ha="center", fontsize=7, color=GR)
plt.tight_layout(pad=0.3)
plt.savefig("images/mcmc_paso5.png", dpi=220, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()

# ══════════════════════════════════════════════════════════
# PASO 6 — convergencia
# ══════════════════════════════════════════════════════════
fig, ax = make_fig()
posterior_rings(ax)
axes_labels(ax)
for (x,y) in cloud:
    c = plt.Circle((x, y), 3, fc=WH, alpha=0.55, ec="none", zorder=5)
    ax.add_patch(c)
ax.text(340, 155, r"nube $\approx p(\theta \mid \mathcal{D})$",
        ha="center", fontsize=8, color=AM, zorder=8)
# Redes de predicción
nets = [(490,100),(520,135),(500,165)]
for nx, ny in nets:
    r = mpatches.FancyBboxPatch((nx-22, ny-12), 44, 24,
                                boxstyle="round,pad=3",
                                fc="#1A0A05", ec=OR, lw=0.8, zorder=6)
    ax.add_patch(r)
    ax.text(nx, ny+2, r"red $\theta^{(s)}$",
            ha="center", fontsize=6.5, color=OR, zorder=7)
    ax.plot([nx-22, 340], [ny, 200], color=WH,
            lw=0.5, ls="--", alpha=0.3, zorder=4)
# Caja predicción
pr = mpatches.FancyBboxPatch((450, 230), 120, 36,
                             boxstyle="round,pad=3",
                             fc="#0A1020", ec=BL, lw=1.0, zorder=6)
ax.add_patch(pr)
ax.text(510, 245, "predicción media",
        ha="center", fontsize=7, color=BL, zorder=7)
ax.text(510, 258, "+ incertidumbre",
        ha="center", fontsize=6.5, color=GR, zorder=7)
step_title(ax, "6. Convergencia y predicción")
ax.text(340, 360, r"Promedio sobre $\{p(y^*|x^*, \theta^{(s)})\}_{s=1}^S$",
        ha="center", fontsize=7, color=GR)
plt.tight_layout(pad=0.3)
plt.savefig("images/mcmc_paso6.png", dpi=220, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()
```
